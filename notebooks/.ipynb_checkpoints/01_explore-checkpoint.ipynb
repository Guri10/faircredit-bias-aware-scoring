```python
import pandas as pd
import matplotlib.pyplot as plt

col_names = [
    "Status_Checking_Account",  # A11–A14
    "Duration",                 # months
    "Credit_History",           # A30–A34
    "Purpose",                  # A40–A410
    "Credit_Amount",
    "Savings_Account",          # A61–A65
    "Present_Employment",       # A71–A75
    "Installment_Rate",         # %
    "Personal_Status_Sex",      # A91–A95
    "Other_Debtors",            # A101–A103
    "Present_Residence_Since",  # years
    "Property",                 # A121–A124
    "Age",                      # years
    "Other_Installment_Plans",  # A141–A143
    "Housing",                  # A151–A153
    "Number_of_Existing_Credits",
    "Job",                      # A171–A174
    "Number_of_People_Liable",
    "Telephone",                # A191–A192
    "Foreign_Worker",           # A201–A202
    "Risk"                      # 1=good, 2=bad
]
```

```python
# Cell 2: Load the true space-delimited file
df = pd.read_csv(
    "../data/raw/german.data",
    sep=r"\s+",       # split on any whitespace
    header=None,      # the file has no header row
    names=col_names,  # your 21 column names list
    engine="python"
)

# Map numeric class → text
df["Risk"] = df["Risk"].map({1: "good", 2: "bad"})
```

```python
print("Shape:", df.shape)
print("\nData types:\n", df.dtypes)
```

```python
counts = df["Risk"].value_counts()
counts.plot(kind="bar")
plt.title("Distribution of Good vs. Bad Borrowers")
plt.xlabel("Risk")
plt.ylabel("Count")
plt.show()
```

```python
num_cols = df.select_dtypes(include=["int64","float64"]).columns.tolist()
df[num_cols].hist(figsize=(12, 8), bins=20)
plt.tight_layout()
plt.show()
```

```python
cat_cols = df.select_dtypes(include=["object"]).columns.drop("Risk")
fig, axes = plt.subplots(len(cat_cols), 1, figsize=(8, 4*len(cat_cols)))
for ax, col in zip(axes, cat_cols):
    df[col].value_counts().plot(kind="bar", ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```
